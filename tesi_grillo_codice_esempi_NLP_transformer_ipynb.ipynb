## Gradio


```python
# in italiano
sentiment = pipeline("text-classification", model="MilaNLProc/feel-it-italian-sentiment")
def sentiment_analysis(text):
    results = sentiment(text)
    return results[0]["label"], round(results[0]["score"], 5)
```

```python
samples = [
    ["La Roma ha vinto la Conference Leangue con un gol di Zaniolo "],
    ["La guerra porta sempre orrori e distruzioni"],
    ["Sono andato in vacanza a Parigi"]]
```

```python
gradio_ui = gr.Interface(
    fn=sentiment_analysis,
    title="Text classification",
    description="Valutazione del tono (positivo o negativo) di un testo",
    inputs=gr.inputs.Textbox(lines=10, label="Inserire qui il testo"),
    outputs=[
        gr.outputs.Textbox(label="Sentiment Label"),
        gr.outputs.Textbox(label="Sentiment Score"),
    ],
    examples = samples
)
```

```python
gradio_ui.launch()
```

## Gradio


```python
model = pipeline("text-generation", model="GroNLP/gpt2-small-italian")

def predict(prompt):
    completion = model(prompt)[0]["generated_text"]
    return completion
```

```python
samples = [
    ["Giovanni entrò in casa sbattendo la porta"],
    ["Era una notte buia e tempestosa"],
    ["C'era una volta un bellissimo cavaliere di nome Rolando"]
]
```

```python
gr.Interface(fn=predict, inputs="text", outputs="text", examples = samples).launch()
```

##Gradio


```python
pipeline_summ = pipeline(
    "summarization",
    model="ARTeLab/mbart-summarization-fanpage"
)

def summarizer(text):
    results = pipeline_summ(text)
    return results[0]["summary_text"]

summary = gr.Interface(
    fn=summarizer,
    inputs=gr.inputs.Textbox(),
    outputs=gr.outputs.Textbox(label="Sintesi del testo"),
)
```
